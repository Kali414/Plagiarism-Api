```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,ReLU,Dropout,LSTM,Embedding,BatchNormalization,Bidirectional
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
```

```python
vocab_size=415
embedding_dims=300
sequence_length=200
dropout_rate=0.3
learning_rate = 0.001   
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
pad_sequenced=pad_sequences(sequences, maxlen=sequence_length, padding='post')
```

```python
pad_sequenced
```

```python
# Model Definition
# Model Definition
model = Sequential()

# Embedding Layer
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dims, input_length=sequence_length))

# LSTM Layers with Batch Normalization and Dropout
model.add(Bidirectional(LSTM(64, activation="relu", return_sequences=True, kernel_regularizer='l2')))
model.add(BatchNormalization())
model.add(Dropout(dropout_rate))

model.add(Bidirectional(LSTM(128, activation="relu", return_sequences=True, kernel_regularizer='l2')))  # Change to return_sequences=False
model.add(BatchNormalization())
model.add(Dropout(dropout_rate))

model.add(Bidirectional(LSTM(256, activation="relu", return_sequences=True, kernel_regularizer='l2')))
model.add(BatchNormalization())
model.add(Dropout(dropout_rate))

model.add(Bidirectional(LSTM(512, activation="relu", return_sequences=False, kernel_regularizer='l2')))
model.add(BatchNormalization())
model.add(Dropout(dropout_rate))

# Dense Output Layer
model.add(Dense(units=512,activation="relu"))
model.add(Dropout(dropout_rate))
model.add(Dense(units=128,activation="relu"))
model.add(Dropout(dropout_rate))
# Dense Output Layer
model.add(Dense(units=64,activation="relu"))
model.add(Dropout(dropout_rate))

model.add(Dense(units=3, activation="softmax"))
```

```python
model.summary()
```

```python
early_stop=tensorflow.keras.callbacks.EarlyStopping(
     monitor='val_loss',  # Monitors validation loss
    patience=3,          # Stops training after 3 epochs with no improvement
    restore_best_weights=True
)
```

```python
optimizer = Adam(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"])

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(pad_sequenced,label,random_state=0,test_size=0.3)
```

```python
from tensorflow.keras.utils import to_categorical

# Assuming y_train and y_test are integers representing class labels
y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)
```

```python
y_train
```

```python
X_train
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=15,batch_size=32,callbacks=[early_stop])
```
